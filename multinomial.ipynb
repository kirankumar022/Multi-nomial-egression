```python
### Multinomial Regression ####
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
le=LabelEncoder()
lb=OneHotEncoder()
mode = pd.read_csv("E:/Assignments/ASsignment week 12/Multi nomial egression/Assignments/mdata.csv")
mode.head(10)
mode['female']=le.fit_transform(mode['female'])
mode['schtyp']=le.fit_transform(mode['schtyp'])
mode['ses']=le.fit_transform(mode['ses'])
mode['prog']=le.fit_transform(mode['prog'])
mode['honors']=le.fit_transform(mode['honors'])

# Correlation values between each independent features
mode.corr()
train, test = train_test_split(mode, test_size = 0.3)
```

```python
import seaborn as sns 

sns.pairplot(mode.iloc[2:11],height=1.0,aspect=0.7)
```

```python
# multinomial option is supported only by the lbfgs and newton-cg solvers
model = LogisticRegression(multi_class = "multinomial",solver = "newton-cg",max_iter=100).fit(train.iloc[:, 2:], train.iloc[:, 10])
```

```python
test_predict = model.predict(test.iloc[:, 2:]) # Test predictions
# Test accuracy 
accuracy_score(test.iloc[:,10], test_predict)
```

```python
train_predict = model.predict(train.iloc[:, 2:]) # Train predictions 
# Train accuracy 
accuracy_score(train.iloc[:,10], train_predict) 
```
